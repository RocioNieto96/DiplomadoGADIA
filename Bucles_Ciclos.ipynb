#FOR
Bucle for El bucle for se utiliza para iterar sobre una secuencia (como una
lista, tupla, diccionario, conjunto o cadena de caracteres). Se ejecuta una vez
por cada elemento en la secuencia.

```python
#imprimir todos lo nmeros de una secuencia de datos
secuencia=[1,2,3,4,5]
for i in secuencia :
  print(i)
```

```python
#clasificar los nmeros que son pares y los impares de la lista secuencia
for i in secuencia  :
  if i%2 == 0 :
    print(f"El nmero {i} es par ")
  else :
    print(f"El nmero {i} es impar")
```

```python
"""Supongamos que quieres solicitar a un usuario que ingrese una contrasea
hasta que lo haga correctamente"""
""" break: Sale del bucle antes de que se evale la condicin o se termine de iterar sobre la secuencia.
continue: Salta el resto del cdigo dentro del bucle y vuelve a la condicin o a la siguiente iteracin.
else: Se ejecuta una vez cuando el bucle termina, a menos que se haya interrumpido con break."""

contrasea_correcta = "segura123"


while True:
    contrasea_ingresada = input("Ingresa la contrasea: ")

    if contrasea_ingresada == contrasea_correcta:
        print("Contrasea correcta. Bienvenido!")
        break  # Sale del bucle si la contrasea es correcta
    else:
        print("Contrasea incorrecta, intenta de nuevo.")
        continue  # Vuelve a pedir la contrasea

else:
    print("Este mensaje solo se ejecutar si el bucle termina sin 'break'.")

```

```python
contrasea_correcta = "segura123"
intentos = 3

while intentos >0 :
    contrasea_ingresada = input("Ingresa la contrasea: ")

    if contrasea_ingresada == contrasea_correcta:
        print("Contrasea correcta. Bienvenido!")
        break  # Sale del bucle si la contrasea es correcta
    else:
        intentos -=1
        print(f"Contrasea incorrecta, te quedan {intentos} intentos.")
    if intentos == 0:
        print("ya no quedan intentos")
```
